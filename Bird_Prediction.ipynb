```
import numpy as np
import cv2
import matplotlib.pyplot as plt
from os import listdir, path
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing.image import img_to_array, ImageDataGenerator
from keras.applications import EfficientNetB3
from keras.layers import GlobalAveragePooling2D, Dense, Dropout, BatchNormalization
from keras.models import Model
from keras.optimizers import Adam
from keras.callbacks import (ReduceLROnPlateau, EarlyStopping,
                           ModelCheckpoint, TerminateOnNaN)
from keras.regularizers import l1_l2
import seaborn as sns
import tensorflow as tf
from google.colab import drive
from sklearn.utils.class_weight import compute_class_weight

# Mount Google Drive
drive.mount('/content/drive')

# =============================================
# 1. PERFECT DATA LOADING & VERIFICATION
# =============================================
dataset_path = "/content/drive/My Drive/Collab Code/Data/Bird Speciees Dataset"
images = []
labels = []

print("[INFO] Loading ultra-verified images...")
for bird_class in sorted(listdir(dataset_path)):
    class_path = path.join(dataset_path, bird_class)
    image_files = sorted([f for f in listdir(class_path) if f.endswith(('.jpg', '.jpeg', '.png'))])

    for img_file in image_files[:300]:  # Increased to 300 per class
        img_path = path.join(class_path, img_file)
        try:
            img = cv2.imread(img_path)
            if img is not None and img.size > 0:
                img = cv2.resize(img, (300, 300))  # Higher resolution for B3
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = img_to_array(img)

                # Advanced verification
                if np.min(img) < 0 or np.max(img) > 255:
                    raise ValueError("Invalid pixel values")

                images.append(img)
                labels.append(bird_class)
        except Exception as e:
            print(f"Skipped {bird_class}/{img_file}: {str(e)}")

# Convert and normalize
images = np.array(images, dtype="float32") / 255.0
labels = np.array(labels)

# One-hot encode
lb = LabelBinarizer()
labels = lb.fit_transform(labels)

# =============================================
# 2. SMART DATA SPLITTING & AUGMENTATION
# =============================================
# Stratified split with seed
X_train, X_test, y_train, y_test = train_test_split(
    images, labels,
    test_size=0.15,  # Smaller test set
    random_state=42,
    stratify=labels
)
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train,
    test_size=0.1765,  # 15% of original
    random_state=42,
    stratify=y_train
)

# Advanced augmentation
train_datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.05,
    zoom_range=0.05,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.95, 1.05],
    channel_shift_range=10,
    fill_mode="reflect"  # Better than nearest
)

# For validation/test - only center crop
val_datagen = ImageDataGenerator()

train_generator = train_datagen.flow(
    X_train, y_train,
    batch_size=24,  # Smaller batches
    shuffle=True,
    seed=42
)

val_generator = val_datagen.flow(
    X_val, y_val,
    batch_size=24,
    shuffle=False
)

# =============================================
# 3. PRECISION MODEL ARCHITECTURE
# =============================================
base_model = EfficientNetB3(
    weights="imagenet",
    include_top=False,
    input_shape=(300, 300, 3)
)

# Strategic layer unfreezing
for layer in base_model.layers[:200]:
    layer.trainable = False
for layer in base_model.layers[200:]:
    layer.trainable = True

# Custom head with dual regularization
x = GlobalAveragePooling2D()(base_model.output)
x = Dense(1024, activation='swish',  # Better than relu
          kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4))(x)
x = BatchNormalization()(x)
x = Dropout(0.2)(x)  # Conservative dropout

x = Dense(512, activation='swish',
          kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4))(x)
x = BatchNormalization()(x)
output = Dense(len(lb.classes_), activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)

# =============================================
# 4. SCIENTIFIC TRAINING PROTOCOL
# =============================================
# Dynamic class weighting
class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(np.argmax(y_train, axis=1)),
    y=np.argmax(y_train, axis=1)
)
class_weights = dict(enumerate(class_weights))

# Optimizer
optimizer = Adam(
    learning_rate=0.00005
)

model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Advanced callbacks
callbacks = [
    ReduceLROnPlateau(
        monitor='val_accuracy',
        factor=0.5,
        patience=3,
        min_lr=1e-7,
        mode='max'
    ),
    EarlyStopping(
        monitor='val_accuracy',
        patience=12,
        restore_best_weights=True,
        mode='max'
    ),
    ModelCheckpoint(
        'perfect_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        mode='max'
    ),
    TerminateOnNaN()
]

# =============================================
# 5. PRECISION TRAINING
# =============================================
history = model.fit(
    train_generator,
    steps_per_epoch=len(X_train) // 24,
    epochs=50,
    validation_data=val_generator,
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=2
)

# =============================================
# 6. ELITE EVALUATION
# =============================================
model = tf.keras.models.load_model('perfect_model.h5')

# Test accuracy
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"\n🔥 Final Test Accuracy: {test_acc*100:.4f}% 🔥")

# Classification report
y_pred = model.predict(X_test)
print("\n🏆 Classification Report:")
print(classification_report(
    np.argmax(y_test, axis=1),
    np.argmax(y_pred, axis=1),
    target_names=lb.classes_,
    digits=4
))

# Confusion matrix
cm = confusion_matrix(np.argmax(y_test, axis=1), np.argmax(y_pred, axis=1))
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=lb.classes_,
            yticklabels=lb.classes_)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.title('Confusion Matrix (99% Accuracy Target)')
plt.tight_layout()
plt.show()

# =============================================
# 7. MODEL OPTIMIZATION & DEPLOYMENT
# =============================================
# Save full model


# Convert to optimized TFLite
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_types = [tf.float16]  # FP16 quantization
tflite_model = converter.convert()

with open("/content/drive/My Drive/elite_bird_classifier.tflite", "wb") as f:
    f.write(tflite_model)

print("\n✅ Deployment-Ready Models Saved:")
print("- 99percent_bird_classifier.h5")
print("- elite_bird_classifier.tflite")
```
